```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf


days = 20
tickers = {
    "apple": "AAPL",
    "microsoft": "MSFT",
    "amazon": "AMZN",
    "meta" : "META",
    "google": "GOOGL",
    "netflix": "NFLX",
}

def get_stock_change(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')

        # 確認するためのprint文を追加
        print(f"Type of index for {company}: {type(hist.index)}")

        # hist.indexがDatetimeIndexである場合のみstrftimeを適用
        if isinstance(hist.index, pd.DatetimeIndex):
            hist.index = hist.index.strftime('%d %B %Y')
        else:
            # エラーを避けるために、インデックスがDatetimeIndexでない場合の処理を記述
            print(f"Index for {company} is not a DatetimeIndex. Skipping strftime.")
        
        # 感想 : そっか、もうFacebookは社名変更してるから、FBでエラーが出るのか。なぜ気づかなかったｗ
        # https://www.click-sec.com/corp/news/info/20211029-01/
        # meta : MVRS ね、理解しました。-> SBI証券の米国株一覧には、metaがない。yfinanceもmetaに対応していない。-> なので、metaは削除。
        # と思ったら、metaはTicker:METAでした、上記情報は一体なんだったんだ。
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        hist.index = [company]
        df = pd.concat([df, hist],axis=0)
    return df
    
df = get_stock_change(days, tickers)
# prophet 使って株価予想も組み込みたいな、一旦完成させてから組み込むか。
df
```

### 以下では、SBI証券の「米国株式取扱銘柄一覧」から、米国株のティッカーと、銘柄名をを取得。

```python
import sqlite3
import requests
from bs4 import BeautifulSoup
```

```python
# スクレイピングするURL
url = 'https://search.sbisec.co.jp/v2/popwin/info/stock/pop6040_usequity_list.html'

# requestsを使ってウェブページを取得する
response = requests.get(url)

# エンコーディングを手動で設定
# 感想 : なんで shift_jis やねん、utf-8 かと思って一時間ぐらい溶かしたわ
response.encoding = 'shift_jis'

# レスポンスのテキストを取得
html = response.text

# BeautifulSoupオブジェクトを作成
soup = BeautifulSoup(html, 'html.parser')

# クラス名に基づいて特定のテーブルを見つける
# 感想 : ページ内に不要なものも含めてテーブルたくさんあるので、クラス指定するのが良いと判断
table = soup.find('table', class_='md-l-table-01 md-l-utl-mt10')

# テーブルが正しく表示されるか確認
# 感想 : OK、これでいけそう
print(table)
```
