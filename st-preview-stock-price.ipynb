```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf


days = 20
tickers = {
    "apple": "AAPL",
    "microsoft": "MSFT",
    "amazon": "AMZN",
    "meta" : "META",
    "google": "GOOGL",
    "netflix": "NFLX",
}

def get_stock_change(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')

        # print
        print(f"Type of index for {company}: {type(hist.index)}")

        # hist.indexDatetimeIndexstrftime
        if isinstance(hist.index, pd.DatetimeIndex):
            hist.index = hist.index.strftime('%d %B %Y')
        else:
            # DatetimeIndex
            print(f"Index for {company} is not a DatetimeIndex. Skipping strftime.")
        
        #  : FacebookFB
        # https://www.click-sec.com/corp/news/info/20211029-01/
        # meta : MVRS -> SBImetayfinancemeta-> meta
        # metaTicker:META
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        hist.index = [company]
        df = pd.concat([df, hist],axis=0)
    return df
    
df = get_stock_change(days, tickers)
# prophet 
df
```
